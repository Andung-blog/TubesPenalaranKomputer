```python
import os
import fitz  # PyMuPDF
import pandas as pd
import re

# Path ke folder berisi semua PDF
pdf_folder = "/content/drive/MyDrive/MA_Putusan/PDF"

# Fungsi bantu untuk ambil teks dari PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Fungsi ekstraksi pola dari teks
def extract_from_text(text):
    def find(pattern, default="Tidak disebutkan", flags=re.IGNORECASE):
        match = re.search(pattern, text, flags)
        return match.group(1).strip() if match else default

    nomor = find(r"PUTUSAN\s+Nomor\s*:\s*(\S+)")
    tanggal = find(r"Tanggal\s+(\d{1,2}\s\w+\s\d{4})|pada hari.*?tanggal\s+(\d{1,2}\s\w+\s\d{4})")
    pihak = find(r"Terdakwa(?:\s*:\s*|\s+)([A-Z\s\.\']{3,})")
    pasal = "; ".join(sorted(set(re.findall(r"Pasal\s+\d+[^\n\s]*", text, re.IGNORECASE))))
    ringkasan_fakta = find(r"MENIMBANG.*?bahwa\s+(.*?)\s+(yang|dimaksud|telah)", default="Tidak ditemukan", flags=re.IGNORECASE | re.DOTALL)

    return {
        "no_perkara": nomor,
        "tanggal": tanggal,
        "ringkasan_fakta": ringkasan_fakta.replace('\n', ' ').strip(),
        "pasal": pasal,
        "pihak": pihak,
        "text_full": text
    }

# Loop semua PDF dalam folder
data = []
pdf_files = sorted([f for f in os.listdir(pdf_folder) if f.endswith(".pdf")])

for i, file_name in enumerate(pdf_files, start=1):
    file_path = os.path.join(pdf_folder, file_name)
    print(f"[{i}] Memproses: {file_name}")

    try:
        text = extract_text_from_pdf(file_path)
        extracted = extract_from_text(text)
        extracted["case_id"] = i
        data.append(extracted)
    except Exception as e:
        print(f"Gagal memproses {file_name}: {e}")

# Buat DataFrame
df = pd.DataFrame(data, columns=[
    "case_id", "no_perkara", "tanggal",
    "ringkasan_fakta", "pasal", "pihak", "text_full"
])

# Simpan ke CSV dan JSON
output_dir = "/content/drive/MyDrive/MA_Putusan/CSV"
os.makedirs(output_dir, exist_ok=True)

csv_path = os.path.join(output_dir, "dataset_putusan_final.csv")
json_path = os.path.join(output_dir, "dataset_putusan_final.json")

df.to_csv(csv_path, index=False)
df.to_json(json_path, orient="records", indent=2, force_ascii=False)

print(f"✅ Dataset berhasil disimpan:\nCSV : {csv_path}\nJSON: {json_path}")
```
